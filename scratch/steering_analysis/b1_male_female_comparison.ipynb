## Helper functions

```python
def plot_spike_rate(ax, data_folder, axo_num, sex, sex_color_flag=True):
    """
    Simple function to make it easier to plot multiple spike rates
    
    """
    # load data 
    data = load_processed_data(data_folder, axo_num)
    if sex_color_flag and (sex == 'F'):
        color = 'lightskyblue'
    elif sex_color_flag and (sex == 'M'):
        color = 'coral'
    else:
        color = 'w'
        
    # plot spike rate
    ax.plot(data['time'], data['spike_rate'], color=color)

    return ax
    
```

## Load data

```python
# Which data files to look at
data_folders = ['84_20250313', '83_20250310']
axo_nums = [2, 2]
sexes = ['M', 'F'] 
```

```python
# make plots showign comparison
fig, ax = plt.subplots(figsize=(12,4))

for ith, (data_folder, axo_num) in enumerate(zip(data_folders, axo_nums)):
    plot_spike_rate(ax, data_folder, axo_num, sexes[ith])
    
```
