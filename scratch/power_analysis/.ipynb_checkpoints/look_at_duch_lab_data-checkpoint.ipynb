## Notebook to look at data from Hurkey, et al (Duch Lab CPG paper)

## Imports

```python
import os
import pyabf 
import numpy as np
import matplotlib.pyplot as plt

from neo.io import Spike2IO
from mosquito.util import butter_bandpass_filter, butter_highpass_filter
```

## Params

```python
plt.style.use('dark_background')
plt.rc('axes', titlesize=18)     # fontsize of the axes title
plt.rc('axes', labelsize=16)    # fontsize of the x and y labels
```

## Load data

```python
# path to data 
data_path = '/media/sam/SamData/Hurkey_et_al_data/Data Figure 1/Fig 1/Panel A+G'
data_fn = 'figure 1a_original recording.smr'
```

```python
# load data file
data = Spike2IO(os.path.join(data_path, data_fn)).read()
```

```python
signal = data[0].segments[0].analogsignals[0]
fs = data[0].segments[0].analogsignals[-1].sampling_rate.magnitude.item()
fs
```

```python
# get trace
sig = signal.magnitude[:,0]
sig
```

## Plot traces

```python
# make figure
fig, ax = plt.subplots(figsize=(11,4))

# set x bounds
t = (1/fs)*np.arange(signal.shape[0])
tmin = 0
tmax = 5   # 0.25
mask = (t >= tmin) & (t <= tmax)

# plot
ax.plot(t[mask], sig[mask]) 

# label axes
ax.set_xlabel('time (s)')
ax.set_ylabel('emg (V)')
```

## Does our Drosophila data look similar?

```python
#load some drosophila data
data_root = '/media/sam/SamData/Mosquitoes'
data_folder = '18_20240508' # '05_20240322'  #'08_20240401' # '04_20240321' #   '05_20240322'  # '03_20240319' # 
axo_num = 3

data_path = os.path.join(data_root, data_folder,
                                 'axo_recording_{:02d}'.format(axo_num))
abf_name = [fn for fn in os.listdir(data_path) if fn.endswith('.abf')][0]
abf_path = os.path.join(data_path, abf_name)

print(abf_path)
```

```python
# load abf file
abf = pyabf.ABF(abf_path)
abf
```

```python
# read out abf fields
# grab data from abf
Fs = abf.sampleRate  # sampling frequency
channel_names = abf.adcNames

# channel data
abf_dict = dict()

for ith, name in enumerate(channel_names):
    # set channel
    abf.setSweep(sweepNumber=0, channel=ith)
    
    # get time once
    if ith == 0:
        abf_dict['Time'] = abf.sweepX
        
    # get current channel data
    abf_dict[name] = abf.sweepY
```

```python
# get emg data
emg = abf_dict['EMG']
```

```python
# filter emg with highpass filter
emg_filt = butter_bandpass_filter(emg, 100, 10000, 1/Fs) 
```

```python
# plot results
# make figure
fig, ax = plt.subplots(figsize=(11,4))

# set x bounds
t = (1/Fs)*np.arange(emg.shape[0])
tmin = 0.08  # 12.5  # 0.08  # 1.185
tmax = 0.12 # 12.75 # 0.12 # 1.195
mask = (t >= tmin) & (t <= tmax)

# plot
ax.plot(t[mask], emg[mask]) 
# ax.plot(t[mask], emg_filt[mask]) 

# label axes
ax.set_xlabel('time (s)')
ax.set_ylabel('emg (V)')
```
