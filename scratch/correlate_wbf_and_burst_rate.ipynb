# Notebook to get correlation coefficient value for WBF and burst rate

The update is that we can now shift in time to maximize the R value

## Imports

```python
import os 
import glob
import pickle

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy import signal

from mosquito.process_abf import load_processed_data
```

## Load data

```python
# look at example data set 
data_folder ='22_20240516'  # '29_20240530'  # '28_20240529'
axo_num =  12  #  12  # 0  # 7  # 8

data = load_processed_data(data_folder, axo_num)
```

```python
# read out some data
wbf = data['wbf']
t = data['time']
spike_rate = data['spike_rate']
spike_idx = data['spike_idx'] 
flying_idx = data['flying_idx']

window = data['params']['emg_window']
fs = data['sampling_freq']
```

## Do cross correlation to get shift that best aligns data

```python
# first, z-score signals
mode='same'
sig1 = wbf[flying_idx]
sig1 -= np.mean(sig1)
sig1 /= np.std(sig1)
sig2 = spike_rate[flying_idx]
sig2 -= np.mean(sig2)
sig2 /= np.std(sig2)

# then take cross corr
corr = signal.correlate(sig1, sig2, mode=mode)
lags = signal.correlation_lags(len(sig1), len(sig2), mode=mode)

# get optimal shift
time_series_shift = lags[np.argmax(corr)]
time_series_shift_sec = time_series_shift/fs
print(f'optimal shift is {time_series_shift_sec} seconds')
# visualize
fig, ax = plt.subplots()

ax.plot(lags, corr)
ax.axvline(0, color='r')

ax.set_title('cross correlation')
ax.set_xlabel('lag')
ax.set_ylabel('correlation')
```

## Get correlation coefficient with shifted values

```python
# correlation without shift:
print('correlation without shift ', np.corrcoef(spike_rate, wbf)[0,1])

# correlation with shift:
shifted_spike_rate = np.roll(spike_rate, time_series_shift)
print('correlation with shift ', np.corrcoef(shifted_spike_rate, wbf)[0,1])
```

## Do linear fit and extract slopes

```python
# linear fit on unshifted data
m,b = np.polyfit(wbf, spike_rate, 1)
print(f'without shift, spike rate = {m} * wbf + {b}')

# linear fit on shifted data
m_shift,b_shift = np.polyfit(wbf, shifted_spike_rate, 1)
print(f'with shift, spike rate = {m_shift} * wbf + {b_shift}')
```
