## Load data

```python
# path to data 
db_path = '/media/sam/SamData/AlyshaEphys/FlyDB'
fly_num = 1525

data_folder = os.path.join(db_path, f'Fly{fly_num:04d}')
```

```python
# get data into a dictionary
data = dict()

# can get more of these later
filename_keys = ['times', 'electrode', 'kinefly_right', 'kinefly_left']

for key in filename_keys:
    with h5py.File(os.path.join(data_folder, f'abf_{key}.hdf5'), 'r') as f:
        data[key] = f[f'abf_{key}'][:]
```

```python
# test that we got it
fig, ax = plt.subplots(figsize=(12,4))

ax.plot(data['times'], data['electrode'])

ax.set_xlim([32.5, 33])
```
